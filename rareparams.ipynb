```python
import os, codecs, traceback
from utils import get_files_ext
from h2p import get_h2p_param
from collections import Counter
```

```python
preset_files = get_files_ext("/Users/naotake/Datasets/diva", "h2p")
```

```python
p_count = Counter()
for pf in preset_files:
    with codecs.open(pf, encoding="ISO-8859-1") as f:
        try:
            param = get_h2p_param(f, "Diva")
        except Exception as e:
            print("{0}, {1}".format(pf, e))
        else:
            p_count.update(list(param.keys()))
```

```python
p_count.most_common()[0][1]
```

```python
rare_params = set()
for pname, v in p_count.items():
    if v< p_count.most_common()[0][1]:
        print(pname, v)
        rare_params.add(pname)
```
